### OS version check

````python
##  OS verison check 
def find_ostype(osversion):
    if "iOS" in osversion:
        return 'iOS'
   else:
        return 'Android'
````

````python
find_ostype("Android7.0")
````

````python
def find_osversion(osversion):
    return osversion.replace("iOS", "").replace("Android", "")
````

````python
find_osversion("Android7.0")
````

### Cleaning devicemanufacturer 

````python
# Cleaning devicemanufacturer
def clean_devicemanufacturer(devicemanufacturer):
    if "samsung" in devicemanufacturer:
        # Large character
        return "Samsung"

    elif "LGE" in devicemanufacturer:
        return "LG"

    elif "LG Electronics" in devicemanufacturer:
        return "LG"

    elif "Apple" in devicemanufacturer:
        return "Apple"

    else:
        return "Other"
````

````python
raw_log["devicemanufacturer(clean)"] = raw_log["devicemanufacturer"].apply(clean_devicemanufacturer)

print(raw_log.shape)

raw_log["devicemanufacturer(clean)"].value_counts()
````

### Organize channel culumn info

````python
import numpy as np

def clean_channel(channel):
    
    if pd.isnull(channel):
        return np.nan
    
    if "google" in channel:
        return "google"
  
    elif "naver" in channel:
         return "naver"
 
    elif "daum" in channel:
        return "daum"

    elif "apple" in channel:
        return "apple"
 
    elif "WEB" in channel:
        return "web"
 
    else:
        return channel
````

````python
def find_viewcategory(inappeventcategory):

    if pd.isnull(inappeventcategory):
          return np.nan

    # inappeventcategory, split .
    chunk = inappeventcategory.split(".")[0]
    viewcategory = chunk.split("_")[0]
    
    return viewcategory
````

````python
# test viewcategory (.)
find_viewcategory("home.view")
````

````python
# test viewcategory (_)
find_viewcategory("gig_detail.view")
````

````python
def find_viewid(inappeventcategory):
 
    if pd.isnull(inappeventcategory):
        return np.nan

    viewid = inappeventcategory.split(".")[0]
        return viewid
````

````python
# test1 viewid
find_viewid("home.view")
````

````python
# test2 viewid
find_viewid("gig_detail.view")
````

````python
def find_viewaction(inappeventcategory):

    if pd.isnull(inappeventcategory):
        return np.nan
    
    viewaction = inappeventcategory.split(".")[1]
        return viewaction
````

````python
# test viewaction
find_viewaction("home.view")
````

````python
# test2 viewaction
find_viewaction("gig_detail.view")
````
